```python
"""
En caso que sea necesario:
Quitar NaNs
Cambiar tipo de dato
Reestructurar
"""

df = pd.read_csv('Natalidad_02.csv')
df=df.dropna()
df=df.drop(labels=0, axis=0)
df['Hombres']=pd.to_numeric(df['Hombres'].str.replace(',',''))
df['Hombres.1']=pd.to_numeric(df['Hombres.1'].str.replace(',',''))
df['Hombres.2']=pd.to_numeric(df['Hombres.2'].str.replace(',',''))
df['Mujeres']=pd.to_numeric(df['Mujeres'].str.replace(',',''))
df['Mujeres.1']=pd.to_numeric(df['Mujeres.1'].str.replace(',',''))
df['Mujeres.2']=pd.to_numeric(df['Mujeres.2'].str.replace(',',''))
hombres=df['Hombres']+df['Hombres.1']+df['Hombres.2']
mujeres=df['Mujeres']+df['Mujeres.1']+df['Mujeres.2']
df2=pd.DataFrame([])
df2['Entidad federativa']=df['Entidad federativa']
df2['Hombres']=hombres
df2['Mujeres']=mujeres
df2['Total']=df2['Hombres']+df2['Mujeres']
df2=df2.drop(  df2[(df2['Entidad federativa']=='Total') | (df2['Entidad federativa']=='Extranjero') ].index  )
#df2=df2.reset_index()
df2
```

```python
"""
Graficar con Seaborn (pairplot, violinplot, heatmap), utilizando en al menos una gráfica 'hue'.
"""
sns.pairplot(df2)
```

```python
plt.rcParams['figure.figsize'] = [10, 7]
sns.violinplot(data=df2,  x="Total", palette='plasma', scale='area')
```

```python
sns.jointplot(data = df2, x = "Hombres", y = "Mujeres", kind='kde')
```

```python
df3=df2.copy()
df3.set_index('Entidad federativa', inplace=True)
sns.heatmap(df3[['Hombres', 'Mujeres']],annot=True)
```

```python
sns.heatmap(df3, annot=True)
```

```python
# Mapa coroplético de México donde se ilustran el total de nacimientos de hombres en los diferentes estados de Mexico. 
# Los datos contiene la información de los años 2017-2019 


geo_url = r'https://raw.githubusercontent.com/angelnmara/geojson/master/mexicoHigh.json'


m = folium.Map(location=[19.4326, -99.1332], zoom_start= 4, width='100%', height='100%',position='bottomLeft')
folium.Choropleth(
    geo_data=geo_url,
    name='Nacimientos de hombres',
    data=df2,
    columns=['Entidad federativa', 'Hombres'],
    key_on='feature.properties.name',
    fill_color='Blues', 
    fill_opacity = 0.9,
    line_opacity = 0.2, 
    nan_fill_color = 'white',
    line_color="black", 
    line_weight=1,
    legend_name='Nacimientos de hombres por entidad federativa del 2017-2019',
).add_to(m)

folium.LayerControl().add_to(m)
m
```
