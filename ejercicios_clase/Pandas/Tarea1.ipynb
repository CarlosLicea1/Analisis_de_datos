```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('GCB2022v27_MtCO2_flat.csv')


"""
Explorar el dataset
desplegar nombres de columnas
tipo de datos
distribucion de NaNs
"""
#desplegando columnas
print(f'Columnas\n{df.columns}')

#desplegando tipos de datos
print(f'\nTipos de datos\n{df.dtypes}')

#desplehando la distribución de NaNs
print(f'\nDistruibucion de los NaNs\n{df.isna().sum(axis=0)}')
```

```python
"""
¿De cuántos países diferentes hay datos?
"""
#contando el número de paises diferentes de los que hay datos
print(f"Hay datos de {df['Country'].nunique()} paises diferentes")
```

```python
"""
Graficar los valores globales y los de Mexico para: 'Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita'.

"""
dfglobal=df[df['Country']=='Global']
dfmexico=df[df['Country']=='Mexico']

columns=['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

for column in columns:
    
    #especificaciones de las gráficas
    plt.title(column)
    plt.xlabel('Year')
    plt.ylabel(column)
    
    plt.plot(dfglobal['Year'], dfglobal[column], label='Global')
    plt.plot(dfmexico['Year'], dfmexico[column], label='Mexico')
    plt.legend()
    plt.show()
```

```python
"""Extraer datos de los países pertenecientes al G20 y graficar las columnas del punto anterior."""

G20=['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Indonesia', 'Italy',
     'Japan', 'South Korea', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 'Turkey', 'United Kingdom',
     'USA']

listG20=[]
for i in range(0, len(G20)):
    listG20.append(df[df['Country']==G20[i]])
columns=['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

for column in columns:
    #especificaciones de las gráficas
    plt.rcParams['figure.figsize'] = [10, 7]
    plt.title(f'{column} G20 countries comparison')
    plt.xlabel('Year')
    plt.ylabel(column)
    for i in range(0, len(G20)):
        plt.plot(listG20[i]['Year'], listG20[i][column], label=G20[i])
    plt.legend()
    plt.show()
    
```

```python
"""
Cuales son los 3 países que más contaminan en cada uno de los casos?

"""
columns=['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

#concatenamos los data frames del G20
dfG20=pd.concat(listG20, axis=0)

total=dfG20.groupby('Country').sum()
print('Paises que más contaminan')
for column in columns:
    aux=list(total.reset_index().nlargest(3, column)['Country'])
    print(f'{column}:\n', '\t', f'{aux}')
```

```python
""" Considerando solo la emisión de los paises del G20, con que porcentaje contribuye México al Total y Per Capita?"""
porcentajeTotal=total.loc['Mexico', 'Total']/  total['Total'].sum()
porcentajePerCapita=total.loc['Mexico', 'Per Capita']/  total['Per Capita'].sum()

print(f'Mexico contribuye al {porcentajeTotal*100:.3}% del total de las emisiones totales de los paises del G20')
print(f'Mexico contribuye al {porcentajePerCapita*100:.3}% del total de las emisiones Per Capita de los paises del G20')
```

```python
"""
Aplicar merge. 
Se aplicó Merge para agregar una columna acerca de si el Pais estaba en la lista del G20 o no
"""
countries=list(df['Country'].value_counts().index)
G20=['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Indonesia', 'Italy',
     'Japan', 'South Korea', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 'Turkey', 'United Kingdom',
     'USA']
df2=[]
decision={True:'Si', False:'No'}
for i in range(df['Country'].size):
    aux=df['Country'][i]
    #print(aux)
    df2.append([aux,  decision[aux in G20] ])
    #print(df2)
df2=pd.DataFrame(df2, columns=['Country', 'G20'])
pd.merge(df, df2, left_on='Country', right_on='Country')
```
